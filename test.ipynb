## Part 2: Neural Network from Scratch with Matrices and NumPy


```python
sys.path.append('./src/matrices_nn/')
import matrices_nn as mnn
import numpy as np
```

Prepare trainings and test data

```python
training_inputs = np.array([np.array(sample[0]).flatten() for sample in dt.trainingSet])
training_labels = np.array([dt.outputDict[sample[1]] for sample in dt.trainingSet])

test_inputs = np.array([np.array(sample[0]).flatten() for sample in dt.testSet])
test_labels = np.array([dt.outputDict[sample[1]] for sample in dt.testSet])
```

Train the network

```python
matrices_nn = mnn.NeuralNetwork([dt.inputDim, dt.hiddenDim, dt.hiddenDim2, dt.outputDim])
matrices_nn.train(training_inputs, training_labels, epochs=1000, learning_rate=0.1)
```

```python
plt.plot(matrices_nn.costs)
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.title('Cost over time')
plt.show()
```

```python
plt.plot(matrices_nn.accuracies)
plt.xlabel('Epoch')
plt.ylabel('Accuracy %')
plt.title('Accuracy over time')
plt.show()
```

Test the trained network with test data

```python
matrices_nn.print_predictions(test_inputs, test_labels)
print(f"Accuracy: {matrices_nn.accuracy(test_inputs, test_labels)}%")
```
